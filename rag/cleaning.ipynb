```python
from IPython.display import display, Markdown

example = "              Segera miliki produk terbaik kami!!! 😍😍😍                  Klik di sini untuk info          lengkap &amp; harga~ hanya Rp1.000,00 saja!!! [LIMITED OFFER] Résumé tersedia sekarang!     "

ct = cleaning.normalize_punctuation(example)
ct = cleaning.decode_html_entities(ct)
ct = cleaning.remove_special_symbols(ct)
ct = cleaning.remove_emoji(ct)
ct = cleaning.remove_accents(ct)
ct = cleaning.normalize_whitespace(ct)

print("---Raw data---")
display(Markdown(f"{example}"))
print("---Data After cleaning---")
display(Markdown(f"**{ct}**"))
```
